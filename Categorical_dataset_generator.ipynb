```python
dataset_train = pd.read_csv("/Users/pajhaver/Documents/Extra Learning/Kaggle/Getting Started/Housing_prices/train.csv")
dataset_test = pd.read_csv("/Users/pajhaver/Documents/Extra Learning/Kaggle/Getting Started/Housing_prices/test.csv")

```

```python
category_train = ['Neighborhood', 'ExterQual', 'KitchenQual', 'Foundation', 'MSSubClass', 'HeatingQC', 'SaleCondition', 
'Exterior1st', 'Exterior2nd', 'SaleType', 'MSZoning', 'SalePrice','Id']
category_test = ['Neighborhood', 'ExterQual', 'KitchenQual', 'Foundation', 'MSSubClass', 'HeatingQC', 'SaleCondition', 
'Exterior1st', 'Exterior2nd', 'SaleType', 'MSZoning','Id']
```

```python
training_set = dataset_train[category_train]
testing_set = dataset_test[category_test]
print(training_set.shape)
print(testing_set.shape)
```

```python
def count_missing(frame, category, cols_missing):
    missing = frame.isnull().sum()
    print(missing)
    for c in category:
        if frame[c].isnull().any():
            frame[c] = frame[c].fillna('MISSING')
            cols_missing.append(c)
    return cols_missing
cols_missing_testing = []
cols_missing_training = []
cols_missing_training = count_missing(training_set, category_train, cols_missing_training)
cols_missing_testing = count_missing(testing_set, category_test, cols_missing_testing)
print(cols_missing_testing)
```

```python
one_hot_cols = ['ExterQual','KitchenQual','Foundation','HeatingQC','SaleCondition']
training_one_hot = pd.get_dummies(training_set, columns=one_hot_cols, drop_first=True)
print(training_one_hot.shape)

testing_one_hot = pd.get_dummies(testing_set, columns=one_hot_cols, drop_first=True)
print(testing_one_hot.shape)
```

```python
training_extra = set(training_one_hot.columns).difference(set(testing_one_hot))
testing_extra = set(testing_one_hot.columns).difference(set(training_one_hot))
print(training_extra)
print(testing_extra)
```
